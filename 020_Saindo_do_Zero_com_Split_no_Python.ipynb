```python
with open('texto_split.txt', "r", encoding="utf-8") as arquivo:
    texto = arquivo.read()
    print(texto)
```

### Separando as palavras por " "(ESPAÇO)

```python
palavras = texto.split()

for palavra in palavras:
    
    print(palavra)
```

```python
palavras = texto.split()

for palavra in palavras:
    
    if ',' in palavra or ';' in palavra or ':' in palavra or '.' in palavra:
        print(palavra[:-1])
    else:
        print(palavra)
        
```

### Contando o número de ocorrências:

```python
palavras.count('acusação')
```

### Separando as palavras baseado em outros caracteres:

```python
palavras_virg = texto.split(',')

for palavra in palavras_virg:
    
    print(palavra)
    
```

```python
linhas = texto.split('\n')

for linha in linhas:    
    print(palavras_virg.index(linha)+1, ')' , linha)
```

```python
for linha in linhas:
    
    trechos = linha.split(',')
    for trecho in trechos:
        print(trecho)
```
